# LINEAR REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("homeprices.xlsx")
df
```

```python
%matplotlib inline
plt.xlabel('AREA(sq fit)')
plt.ylabel('PRICE(us$)')
plt.scatter(df.area, df.price, color='red', marker='+')

```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python

reg.coef_
```

```python
reg.intercept_
```

```python
# y = mx + c
132.70547945*3300+189726.0273972602
```

```python
d = pd.read_excel("area.xlsx")
d
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_excel("prediction.xlsx")
```

```python
%matplotlib inline
plt.xlabel('AREA(sq fit)')
plt.ylabel('PRICE(us$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
